#### Birleştirme (Join) İşlemleri

```python
import numpy as np
import pandas as pd
```

```python
m = np.random.randint(1,30,size=(5,3))
df1 = pd.DataFrame(m,columns=("var1","var2","var3"))
df1
```

```python
df2 = df1 + 99
df2
```

```python
pd.concat([df1,df2]) # varsayılan olarak axis 0'dır. Yani satır olarak birleştirme yapar
```

```python
?pd.concat
```

```python
pd.concat([df1,df2],ignore_index=True) # ignore index diyerek aslında indexleri
# baştan aşağıya sıralayarak birleştirme işlemini tamamladı.
```

```python
df1.columns = ["var1","var2","deg3"] # df1'in sütunlarından birini değiştirdim.
df1
```

```python
pd.concat([df1,df2]) # Sonuçta görüldüğü gibi ortak olmayan sütunlardaki değerler NaN geldi.
# Bunu aşmanın bazı yolları vardır.
```

```python
pd.concat([df1,df2],join="inner") # sadece iki dataframede de ortak olan sütunları birleştirilecektir.
```
